<h2 style="color:#A80891;font-weight:bold; margin-top:0;">
 Formulate Hypotheses
</h2>

<p style="font-size:14px; color:#444;">
Convert business questions into appropriate statistical tests
to enable data-driven decision making.
</p>

<hr style="border:0.5px solid #ddd;">

<div style="margin-bottom:14px;">
<b>• Gender-based Spending Behavior</b><br>
<span style="color:#555; font-size:14px;">
Business Question: Do males and females spend differently?<br>
Statistical Test: <b>Independent t-test</b>
</span>
</div>

<div style="margin-bottom:14px;">
<b>• Education Level and Monthly Spend</b><br>
<span style="color:#555; font-size:14px;">
Business Question: Does education level impact average monthly spend?<br>
Statistical Test: <b>One-way ANOVA</b>
</span>
</div>

<div style="margin-bottom:14px;">
<b>• Marital Status and Pet Ownership</b><br>
<span style="color:#555; font-size:14px;">
Business Question: Is marital status related to the number of pets owned?
<span style="color:#C0392B;">(Only for DS Students)</span><br>
Statistical Test: <b>Chi-square test</b>
</span>
</div>

<div style="margin-bottom:14px;">
<b>• Age and Activity Level</b><br>
<span style="color:#555; font-size:14px;">
Business Question: Are older people less active?<br>
Statistical Test: <b>Correlation Analysis</b>
(Age vs DaysSinceLastInteraction)
</span>
</div>

<div>
<b>• State-wise Spending Variation</b><br>
<span style="color:#555; font-size:14px;">
Business Question: Does spending vary significantly across states?<br>
Statistical Test: <b>ANOVA</b>
</span>
</div>

</div>


```python
from scipy.stats import ttest_ind, f_oneway, chi2_contingency, pearsonr

#-------------------------------------------------------------------------------
#--------------Independent t_test-------------------------
#-------------------------------------------------------------------------------

# Extract Monthly Spend for Male customers
male_spend = Cust_Data[Cust_Data['Gender'] == 'Male']['MonthlySpend']

# Extract Monthly Spend for Female customers
female_spend = Cust_Data[Cust_Data['Gender'] == 'Female']['MonthlySpend']

# Perform independent two-sample t-test (Welch’s t-test)
# equal_var=False is used when population variances are assumed to be unequal
t_stat, p_value = ttest_ind(
    male_spend, 
    female_spend, 
    equal_var=False
)

# Display t-statistic and p-value
print("Independent t-test: Monthly Spend by Gender")
print(f"T-statistic:{t_stat:.4f}")
print(f"P-value_gender:{p_value:.4f}\n")

#--------------------------------------------------------------------------------------------
#-----------------One-way Anova---------------------------------
#--------------------------------------------------------------------------------------------

# Create a list of MonthlySpend arrays for each Education group
# Each element in the list represents one education category
groups = [
    group['MonthlySpend'].values
    for name, group in Cust_Data.groupby('Education')
]

# Perform one-way ANOVA test
# This checks whether the mean Monthly Spend differs across Education levels
f_stat, p_value = f_oneway(*groups)

# Display F-statistic and p-value
print("One-way ANOVA: Monthly Spend by Education level")
print(f"F-statistic:{f_stat:.4f}")
print(f"P-value_Edu:{p_value:.4f}")

#------------------------------------------------------------------------------------------------------
#--------------Chi-square test of independence---------------------------
#------------------------------------------------------------------------------------------------------

# Create a contingency table between Marital Status and Number of Pets
# Rows represent Married status, columns represent different NumPets categories
contingency = pd.crosstab(
    Cust_Data['Married'],     # Marital status categories
    Cust_Data['NumPets']      # Number of pets categories
)

# Perform Chi-square test of independence
# This tests whether Marital Status and Number of Pets are statistically independent
chi2, p_value, dof, expected = chi2_contingency(contingency)

# Display only the p-value
print("\nChi-square Test: Marital Status vs Number of Pets")
print(f"Chi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p_value:.4f}\n")

#-----------------------------------------------------------------------------------------------------------
# -----------------Correlation analysis-------------------------
# ----------------------------------------------------------------------------------------------------------

#  Calculate Pearson correlation
corr, p_value = pearsonr(
    Cust_Data['Age'],
    Cust_Data['DaysSinceLastInteraction']
)

# Display results
print("Correlation Analysis: Age vs Days Since Last Interaction")
print(f"Correlation coefficient: {corr:.4f}")
print(f"P-value: {p_value:.4f}\n")

#----------------------------------------------------------------------------------------------------------
#------------------One-way ANOVA---------------------
#-----------------------------------------------------------------------------------------------------------

# Create a list of MonthlySpend arrays for each State
# Each array represents MonthlySpend values of customers from one state
state_groups = [
    group['MonthlySpend'].values
    for name, group in Cust_Data.groupby('State')
]

# Perform one-way ANOVA test
# This checks whether the mean Monthly Spend differs significantly across States

f_stat, p_value = f_oneway(*state_groups)

# Display only the p-value
print("One-way ANOVA: Monthly Spend by State")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

#---------------------------------------------------------------------------------------------------------------
```

<h3 style="color:#A80891;font-size:18px;font-weight:bold">
Insights from Hypothesis Testing Analysis
</h3>

<p>
<strong>Spending behavior</strong> remains largely uniform across gender, education levels, and states,
indicating minimal demographic influence on monthly expenditure.
</p>

<h3 style="color:#A80891;font-weight:bold">Understanding Statistical Hypotheses</h3>

<b>Null Hypothesis (H₀)</b>
<p style="font-size:14px; color:#555; margin-top:6px;">
The null hypothesis assumes that there is <b>no effect</b>, <b>no difference</b>,
or <b>no relationship</b> between variables. It represents the
<b>default business assumption</b> and serves as the baseline against which
statistical evidence is evaluated. Statistical tests are designed to assess
whether sufficient evidence exists to <b>reject H₀</b>.
</p>

<b>Alternate Hypothesis (H₁)</b>
<p style="font-size:14px; color:#555; margin-top:6px;">
The alternate hypothesis assumes that a <b>statistically significant effect</b>,
<b>difference</b>, or <b>relationship</b> exists between variables. It represents
the <b>analytical claim</b> and is supported when the null hypothesis is accepted
after rejecting the null hypothesis based on statistical evidence.
</p>
<h4 style="color:#A80891; font-weight:bold;font-size:18px">1. Gender-based Spending Behavior</h4>

<b>Null Hypothesis (H₀):</b>
There is no significant difference in average monthly spend between male and female customers.

<p>
<b>Alternate Hypothesis (H₁):</b>
There is a significant difference in average monthly spend between male and female customers.

<p>
<b>Test Used:</b> Independent t-test
</p>
<p>
<b>Result & Interpretation:</b><br>

This indicates that gender does not significantly influence monthly spending.

<h4 style="color:#A80891; font-weight:bold; font-size:18px;">
Normality Assumption
</h4>

<p>
<b>Purpose:</b><br>
To check whether numerical variables (e.g., MonthlySpend) follow a normal distribution.
</p>

<p>
<b>Method Used:</b><br>
Shapiro–Wilk test<br>
Visual inspection using histogram and KDE plots
</p>



```python
from scipy.stats import shapiro
#-------------------------------------------------------------------
#-----------Shapiro–Wilk test------------
#-------------------------------------------------------------------

# Perform Shapiro–Wilk test for normality on MonthlySpend
# This test checks whether the data follows a normal distribution
shapiro_test = shapiro(Cust_Data['MonthlySpend'])

# Display the test statistic and p-value
print("Assumption of Normality Distribution:")
print(f"Shapiro_Test:{shapiro_test:}")

#----------------------------------------------------------------------
# KDE plot + Histogram 
#----------------------------------------------------------------------

# Create one figure with 2 subplots
plt.figure(figsize=(12,5))

# ---------------- KDE Plot ----------------
plt.subplot(1,2,1)
sns.kdeplot(
    data=Cust_Data,
    x='MonthlySpend',
    hue='Education',
    fill=True
)
plt.title("Spending by Education Level")
plt.xlabel("Monthly Spend")
plt.ylabel("Density")

# ---------------- Histogram ----------------
plt.subplot(1,2,2)
plt.hist(Cust_Data['MonthlySpend'], bins=20, color="purple", edgecolor="black")
plt.xlabel("Monthly Spend")
plt.ylabel("Count")
plt.title("Monthly Spend Distribution")

# Adjust layout
plt.tight_layout()
plt.show()

#---------------------------------------------------------------------------------------
```

<p>
<b>Interpretation:</b><br>
The Shapiro–Wilk test indicated deviation from normality (p &lt; 0.05).
However, given the large sample size (over 10,000 observations),
parametric tests were considered appropriate under the Central Limit Theorem.
</p>

<p>
<b>Conclusion:</b><br>
The normality assumption was reasonably satisfied for hypothesis testing.
</p>
<p>
Normality was assessed using the Shapiro–Wilk test, which indicated deviation
from normality due to the large sample size. Visual inspection through
histograms and KDE plots showed a right-skewed but stable distribution.
Given the large number of observations, parametric tests were applied
under the Central Limit Theorem.
</p>
